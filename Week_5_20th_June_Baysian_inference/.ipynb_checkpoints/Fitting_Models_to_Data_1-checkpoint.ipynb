### $ Y = α + (β \times X) + \sigma $
#### where
#### X : Independent Variable |  Y : Dependent Variable
#### α : Intercept | β : Slope | σ : Intrinsic Scatter

## Let's put it all together

```python
def try_parameters(alpha, beta, sigma):
    
    """Produce Diagnostic Plots for the input parameters
    of the model Y ~ N(α + β*X, σ**2)
    
    Parameters
    ----
    alpha: Intercept
    beta: Slope
    sigma: Intrinsic Scatter
    
    Returns
    ----
    None
    
    Notes
    ----
    σ should be positive
    """
    
    # Receive parameter guesses
    α_try = alpha
    β_try = beta
    σ_try = sigma
    
    # Likelihood
    μ_try =  α_try + β_try * x
    
    # Add the scatter
    y_try = np.random.normal(loc=μ_try, scale=σ_try)
    
    
    # Compare the original 'y' value against the one generated by guessing
    likelihood_proxy = (y - y_try) ** 2
    
    # Plot diagnostics
    datastyle = dict(edgecolor='k', alpha=1.)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 5), 
                                   gridspec_kw={'wspace':0.3})
    sns.scatterplot(x, y, label='Real', ax=ax1, **datastyle)
    sns.scatterplot(x, y_try, label='Tried', ax=ax1, **datastyle)
    ax1.set(
        xlabel='X',
        ylabel='Y',
        title='Comparison of Real Data vs Simulated Data'
    )
    ax1.legend(ncol=2, framealpha=0.6);

    sns.scatterplot(x, likelihood_proxy, color='tab:red', ax=ax2, **datastyle)
    ax2.set(
        xlabel=r'$X$',
        ylabel=r'Likelihood Proxy ($Y - Y_{\mathrm{try}})^{2}$',
        title='Residual Squared Distribution'
    );
    
    return
```

## Doing it with emcee

### Define Prior

```python
def log_prior(theta):
    """
    Sets uniformative prior in the ranges set by us. 
    """
    α, β, log_sigma = theta
    if -10.0 < α < 10. and -10.0 < β < 10.0 and -10.0 < log_sigma < 2.0:
        # This is the log_prior. If parameters are within this range, 
        # then return log_prior = 0. i.e. Prior_prob.  = np.exp(0.) = 1
        # This sets a uniform prior over the above ranges.
        return 0.0
    
    # if the parameters are out of the above ranges, return negative inf
    # this results in zero prior (and therefore, posterior)
    # probability of the parameters.
    return -np.inf
```

### Define Likelihood

```python
def log_likelihood(theta, x, y):
    """Returns Log of Normal Likelihood of our model
    """
    α, β, log_sigma = theta   
    mu = α + β * x
    sigma2 = np.exp(2 * log_sigma) #mu ** 2 * 
    return -0.5 * np.sum((y - mu) ** 2 / sigma2 + np.log(sigma2))
```

### Define Posterior

```python
def log_probability(theta, x, y):
    """Returns the log of posterior probability.
    
    """
    lp = log_prior(theta)
    if not np.isfinite(lp):
        return -np.inf
    # Remember that: Posterior = Prior x Likelihood (upto a constant)
    # Therefore, log_posteriod = log_prior + log_likelihooda
    return lp + log_likelihood(theta, x, y)
```

```python
guesses = [1.5, 2.5, 0.5]
```

```python
import emcee

pos = guesses + 1e-4 * np.random.randn(32, 3)
nwalkers, ndim = pos.shape

sampler = emcee.EnsembleSampler(nwalkers, ndim, log_probability, args=(x, y));
sampler.run_mcmc(pos, 5000, progress=True);
```

```python
fig, axes = plt.subplots(3, figsize=(10, 7), sharex=True)
samples = sampler.get_chain()
labels = ['alpha', 'beta', 'logsigma']
for i in range(ndim):
    ax = axes[i]
    ax.plot(samples[:, :, i], "k", alpha=0.3)
    ax.set_xlim(0, len(samples))
    ax.set_ylabel(labels[i])
    ax.yaxis.set_label_coords(-0.1, 0.5)

axes[-1].set_xlabel("step number");
```

```python
## Discard initial tuning steps and select every 15th step
flat_samples = sampler.get_chain(discard=300, thin=15, flat=True)
```

```python
fig = corner.corner(
    flat_samples, 
    truths=[α_true, β_true, np.log(0.2)], 
    show_titles=True, 
    labels=['alpha', 'beta', 'logsigma']
);
```
